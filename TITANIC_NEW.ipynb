```
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```
df_new['Age']=df_new[["Age",'Pclass']].apply(impute_age,axis=1)
```

```
df_new= df_new.dropna(subset=['Embarked'])
code= pd.get_dummies(df_new['Embarked'], prefix='embarked')
df_new= pd.concat([df_new, code], axis=1)
df_new.head(10)
```
